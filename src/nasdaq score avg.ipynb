```python
import pandas as pd
```

```python
nasdaq = pd.read_csv('/Users/ngocnguyen/Downloads/HistoricalData_1739420282858.csv')
```

```python
nasdaq.head()
```

```python
# Convert 'Date' column to datetime format
nasdaq['Date'] = pd.to_datetime(nasdaq['Date'], errors='coerce')

# Filter for records in the year 2018
nasdaq = nasdaq[(nasdaq['Date'] >= '2018-01-01') & (nasdaq['Date'] <= '2018-12-31')]

# Ensure 'Open' and 'Close/Last' are numeric
nasdaq['Open'] = pd.to_numeric(nasdaq['Open'], errors='coerce')
nasdaq['Close/Last'] = pd.to_numeric(nasdaq['Close/Last'], errors='coerce')

# Calculate avg_value
nasdaq['avg_value'] = (nasdaq['Close/Last'] - nasdaq['Open']) / nasdaq['Open']
```

```python
nasdaq.head(10)
```

```python
nasdaq["Index Price"] = nasdaq["Close/Last"] - nasdaq["Open"]
nasdaq.head()
```

```python
# Drop unnecessary columns
nasdaq = nasdaq.drop(columns=["Open", "High", "Low", "avg_value"])
nasdaq.head()
```

```python
nasdaq.to_csv("nasdaq_index.csv", index=False)
```
